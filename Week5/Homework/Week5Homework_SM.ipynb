```python
import nltk
import pandas as pd
import numpy as np
```

```python
def readFiletoDataFrame(filePath, delimiter):    
    df = pd.read_csv(filePath,sep=delimiter)
    #print(df.head)
    return df

def timeDiff(dataFrame):
    # Adding 2 New Columns to DataFrame. TimeInterval and ActionSequence
    dataFrame["TimeInterval"] = dataFrame["Timestamp"].diff() # Shows the time interval between previous and next timestamp
    #dataFrame["ActionSequence"] = dataFrame["Coding"].values 
    dataFrame["ActionSequence"] = dataFrame.groupby("TestTakerID").cumcount()
    print("Time Interval and Action Sequence of each test taker")
    print(dataFrame)
    #Group By TestTakerID
    #lstGroupBy = list(dataFrame.groupby("TestTakerID"))    
    #print(lstGroupBy)   
    return dataFrame

def firstEvent(dataFrame):
    #filter the dataframe to get the Start events for each Testtaker    
    firstEvents = dataFrame.query("Coding == 'START'") 
    print("First Events of each test taker")
    print(firstEvents)
    return firstEvents

def bigrams(dataFrame,n):
    #flatten the coding values to find bigrams
    #combine events
    bigramList = list(nltk.bigrams(dataFrame["Coding"]))
    
    bigramList.remove(('END', 'START'))
    
    print("bigrams")
    #removes the END from 1 student and Start from other
    bigrams = [x for x in nltk.bigrams(dataFrame["Coding"]) if x != ('END', 'START')]  
    
    print(bigrams)
    bigram_fd = nltk.FreqDist(bigrams)
    #bigram_fd
    
    print(bigram_fd.most_common())
    
    return

def main():    
    filePath = 'C://Users/snmishra/OneDrive - Educational Testing Service/DataScienceAcademy/Week5/Synthetic Dataset/PSTRE_syntheticData.csv'
    delimiter = "\t"
    n = 2
    originalFrame = readFiletoDataFrame(filePath, delimiter)

    timeDiff(originalFrame)
       
    firstEvent(originalFrame)
    
    bigrams(originalFrame, n)
    
    return

main()
```
